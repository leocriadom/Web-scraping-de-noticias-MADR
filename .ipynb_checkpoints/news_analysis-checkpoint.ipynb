### Nube de palabras

```python
t_noticias = news.noticia.str.cat(sep=',') # Un solo string
puntuacion = string.punctuation + '–¡‘’¿”“•\r´'
```

```python
def no_punt(string):
    strip_punt = string.maketrans('', '', puntuacion)
    return string.translate(strip_punt)
```

```python
t_noticias = no_punt(t_noticias)
t_noticias = t_noticias.lower()
len(t_noticias)
```

```python
t_noticias = word_tokenize(t_noticias)
def stopword(tokens):
    return [w for w in tokens if w not in stopwords_esp]
stopwords_esp = stopwords.words('spanish')
t_noticias = stopword(t_noticias)
len(t_noticias)
```

```python
nltk.FreqDist(t_noticias).most_common()
```

```python
def col_counter(lista_palabras, palabra):
    counter = collections.Counter(w for w in lista_palabras if w in palabra)
    return counter[palabra]
```

###  Colocaciones

```python
news_Text = nltk.Text(t_noticias)
news_Text
```

```python
# Bi-gramas alrededor de restitución
bigram_measures = nltk.collocations.BigramAssocMeasures()
creature_filter = lambda *w: 'restitución' not in w
finder = BigramCollocationFinder.from_words(
   news_Text)
finder.apply_freq_filter(5)
finder.apply_ngram_filter(creature_filter)
for i in finder.score_ngrams(bigram_measures.likelihood_ratio):
    print (i)
```

```python
# Tri-gramas alrededor de productivos
trigram_measures = nltk.collocations.TrigramAssocMeasures()
creature_filter = lambda *w: 'productivos' not in w
finder = TrigramCollocationFinder.from_words(
   news_Text)
finder.apply_freq_filter(10)
finder.apply_ngram_filter(creature_filter)
for i in finder.score_ngrams(trigram_measures.likelihood_ratio):
    print (i)
```
